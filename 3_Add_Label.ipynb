# 라벨 붙이기

```
import pandas as pd
```

```
# 자료들 합치기
df1=pd.read_csv('Pre2014_1.csv')
df2=pd.read_csv('Pre2014_2.csv')
df3=pd.read_csv('Pre2015.csv')
df4=pd.read_csv('Pre2016.csv')
df5=pd.read_csv('Pre2017.csv')
df6=pd.read_csv('Pre2018.csv')
df = pd.concat([df1, df2, df3, df4, df5, df6])

Preprocessing = df.preprocessing
```

```
df['Date'] = df['Date'].str[:10] #날짜 정리
```

```
# 라벨 파일 불러온 뒤 빈 값 없애기
label = pd.read_csv('label.csv', dtype=str, sep=',', encoding='utf-8')
label = label.dropna()
```

#날짜를 문자열로 변환 후 merge

```
merge_data = pd.merge(df, label, on="Date", how='left')
```

```
# 라벨이 없는 값은 앞의 라벨 값으로 붙이기
merge_data = merge_data.fillna(method = 'ffill' )
```

```
merge_data.to_csv('KoreanAir_label.csv', sep=',',index=False)
```
