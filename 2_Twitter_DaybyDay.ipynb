# 일별 데이터를 일자별로

https://jsideas.net/daily_to_weekly/

필요한 모듈 import 하기

```python
import pandas as pd
import numpy as np
```

csv 파일 읽어와서 DataFrame으로 만들기

```python
df = pd.read_csv('https://drive.google.com/open?id=1A_HPTD7pLHqtmpWtmowX3Jw-TW8Ablg5')
```

int로 된 날짜를 datetime object로 바꿔주기

```python
df['datetime'] = df['time'].apply(lambda x: pd.to_datetime(str(x),format='%Y-%m-%d'))
```

임의의 값 1을 넣어 DataFrame 완성하기

```python
df['all_tweet_num'] = 1
```

datetime 컬럼을 index로 만들기

```python
print(df.index)
```

```python
df
```

```python
df.set_index(df['datetime'], inplace=True)
```

```python
df = df.drop('datetime', 1)
```

```python
df
```

주별, 월별 데이터로 기준 재정렬하기

```python
daily_df = df.resample('D', how={'all_news_num':np.sum}).fillna(0)
```

```python
print(daily_df)
```

```python
daily_df.to_csv('DailyTweets.csv')
```
