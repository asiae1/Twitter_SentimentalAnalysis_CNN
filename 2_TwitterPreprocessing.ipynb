```python
# 파일 읽어오기
import pandas as pd

df = pd.read_csv('Korean2017_1.csv', dtype=str, sep=',', engine='python', encoding='utf-8')

content = df['Content']
date = df['Date']

df.shape
```

## 명사, 형용사 중 길이 1개 이상만 뽑기


```python
# 토큰화 및 형태소 분석(명사, 형용사, 길이 1개 이상)
from konlpy.tag import Okt
collect_tag = []

okt = Okt()
for sentence in content:
    morph = okt.pos(sentence, norm=True, stem=True)
    list = [n for n, tag in morph if len(n)>1 and tag =='Noun' or tag =='Adjective']
    collect_tag.append(list)
print(collect_tag)
```

```python
# 토큰 모으기
collect_tag2 = []
for i in collect_tag:
  i = (",".join(i))
  collect_tag2.append(i)
print(collect_tag2)

#문장별
data = pd.DataFrame(collect_tag2, columns = ['Token'])
data.shape
data.head()
```

# 2. 불용어 리스트 만들고 제거

```python
# 선별된 품사별 빈도수 계산 & 상위 빈도 10위 까지 출력
from collections import Counter

count_words = []
for i in collect_tag:
  for j in i:
    count_words.append(j)
counts = Counter(count_words)
print(counts.most_common(10))
```

```python
# 최빈어 조회. 최빈어를 조회하여 불용어 제거 대상을 선정
words = Counter(count_words).most_common()
print(words)
```

```python
#104번 이하로 나온 단어 제거(104번부터 쓸모없는 것 같아서)
stopword = []

for i in range(len(words)):
    if words[i][1] <= 104 :
        stopword.append(words[i][0])
print(stopword)
```

```python
#불용어 제거한 리스트
def make_clear_(words):
  make_clear = []
  for word in words:
      if word not in stopword:
          make_clear.append(word)
  return make_clear
      
#print(make_clear)
```

```python
# 원본에서 불용어 지우기
collect_clear = []
result = []

for i in collect_tag:
  for j in i:
    j = (('').join(j))
    collect_clear.append(j)

for word in collect_clear:
  if word not in stopword:
    result.append(word)

print(result)
    
```

```python
result = pd.concat([date, content, data], axis=1)
result.shape
result['preprocessing'] = ""


for i, row in result.iterrows():
  result['preprocessing'][i] = ','.join(make_clear_(row['Token'].split(',')))

result.head()
```

```python
result.to_csv('Pre2017_1.csv', index = False)
```
