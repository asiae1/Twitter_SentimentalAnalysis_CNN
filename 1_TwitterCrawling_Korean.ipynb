```
# 트위터 데이터 크롤링
from twitterscraper import query_tweets
import datetime as dt
from time import sleep
import re

def test(data):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣0-9]+') 
    result = hangul.sub(' ', data) 
    return result

result=[]
if __name__ == '__main__':
    list_of_tweets = query_tweets('대한항공', limit=None, begindate=dt.date(2014,11,16), enddate=dt.date(2015,5,15),  poolsize=20, lang='ko')

    for tweet in list_of_tweets:
        result.append([tweet.timestamp, test(tweet.text)])
result
```

```
#데이터프레임으로 만들고 저장
import pandas as pd

data = pd.DataFrame(result, columns=["Date", "Content"])
result_df = data.drop_duplicates(['Content'], keep='first')
result = result_df.sort_values(by="Date", ascending=False)
result.to_csv('Korean2014_2.csv', index = False)
```
